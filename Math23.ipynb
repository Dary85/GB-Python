```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
ax = plt.gca()

x = np.linspace(-5, 2.0*np.pi, 101)

y1 =2*np.cos(x-3)+4
y2 =5*np.cos(x-1)+2
ax.plot(x, y1)
ax.plot(x, y2)
plt.axis('scaled')
plt.show()
```

```python
p = 3
a =2/3*np.pi
print(f"x = {p*math.cos(a)} y = {p*math.sin(a)}")
```

```python
ax = plt.gca()

t = np.linspace(0, 2 * np.pi, 100)
r = 2
x = r * np.cos(t)
y = r * np.sin(t)

ax.plot(x, y)
plt.axis('scaled')
plt.show()
```

```python
ax = plt.gca()

t = np.linspace(0, 2, 100)

x = np.cos(t)
y = x-3

ax.plot(x, y)
plt.axis('scaled')
plt.show()
```

```python
ax = plt.gca()

x = np.linspace(-5, 5, 50)

y1 = x**2-1 
y2 = (np.exp(x)+1)/x+1

ax.plot(x, y1)
ax.plot(x, y2)

```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D


fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 2)
Y = np.arange(-50, 50, 20)
print(X)
X, Y = np.meshgrid(X, Y)
Z = (10+X) + (10+Y)
Z2 =X + Y
ax.plot_surface(X, Y, Z)
ax.plot_surface(X, Y, Z2)
ax.scatter(0, 0, 0,'z',50,'red')
show()
```

```python
fig = figure()
ax = Axes3D(fig)

X = np.arange(-10, 10, 0.5)
Y = np.arange(-10, 10, 0.5)

X, Y = np.meshgrid(X, Y)
Z =np.sqrt(X**2 + Y**2)


ax.plot_surface(X, Y, Z)

ax.scatter(X, Y, Z,'z')
show()
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x=np.arange(-15,15,0.001)
plt.plot(x,2/(x),'o')
plt.ylabel('y')
plt.xlabel('x')
plt.grid(True)
plt.axis([-20,20,-20,20])
plt.show()
```

```python
ax = plt.gca()

theta = np.linspace(0, 2 * np.pi, 100)
r = 1
x = r * np.cos(theta)
y = r * np.sin(theta)

ax.plot(x, y)
plt.axis('scaled')
plt.show()
```

```python
ax = plt.gca()

theta = np.linspace(0, 2 * np.pi, 100)
r = 1
r1 = 2*r
x = r1 * np.cos(theta)
y = r * np.sin(theta)

ax.plot(x, y)
plt.axis('scaled')
plt.show()
```
